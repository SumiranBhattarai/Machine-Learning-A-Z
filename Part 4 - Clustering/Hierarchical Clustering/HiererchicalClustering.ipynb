# Using the dendogram to find the optimal number of clusters

```python
from scipy.cluster import hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(X,method='ward')) #ward = observation don't vary too much. minimizing variance in clustering
```

```python
from scipy.cluster import hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(X,method='ward')) #ward = observation don't vary too much. minimizing variance in clustering
plt.title('Dendogram')
plt.xlabel('No of Customers')
plt.ylabel('Vertical Distance')
plt.show()
```

# Train the Hiererchical Clustering Algorithm

```python
from sklearn.cluster import AgglomerativeClustering #Classic Hiererchical Clustering
hc= AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(X)
```
